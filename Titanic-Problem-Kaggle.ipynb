```python
train_data = train_data.drop(['Cabin', 'Name'], axis=1)
```

```python
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values=np.nan, strategy='median')
train_data['Age'] = imp.fit_transform(train_data[["Age"]])
```

```python
train_data.isnull().sum()
```

```python
x = train_data.dropna()
```

```python
x.isnull().sum()
```

```python
P_class = pd.get_dummies(x.Pclass, drop_first=True)
Gender = pd.get_dummies(x.Sex, drop_first=True)
Embark = pd.get_dummies(x.Embarked, drop_first=True)
x1 = pd.concat([x, P_class, Gender, Embark], axis=1)
x1.head()
```

```python
x2 = x1.drop(['Pclass', 'Sex', 'Ticket', 'Embarked'], axis=1)
```

```python
x2['male'] = x2['male'].astype(float)
```

```python
x2.head()
```

```python
test_size=0.6
seed = 10
X_train, X_test, Y_train, Y_test = train_test_split(x2.drop("Survived", axis=1), x2['Survived'], test_size=test_size, random_state=seed, stratify=x2["Survived"])
```

```python
model = LogisticRegression(random_state=10, class_weight='balanced', max_iter=1000)
model.fit(X_train,Y_train)
```

```python
prediction= model.predict(X_test)
prediction
```

```python
prediction.reshape(1, -1)
```

```python
model.score(X_test, Y_test)
```
